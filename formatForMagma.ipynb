```python
import pandas as pd
import re
from collections import Counter
import numpy as np
```

```python
def formatSumStats(filepath):
    sumStats = pd.read_csv(filepath, delim_whitespace=True)
    split = sumStats['SNP'].str.split(':', expand=True)
    sumStats.dropna(subset=['OR', 'STAT', 'P'], inplace=True)
    sumStats['A1'] = split[2]
    sumStats['A2'] = split[3]
    sumStats['ID'] = split[4]
    sumStats["CHR"] = sumStats["CHR"].astype(str)
    sumStats["BP"] = sumStats["BP"].astype(int)
    sumStats = sumStats[['ID', 'CHR', 'BP', 'A1', 'A2', 'STAT', 'SE', 'P']]
    sumStats = sumStats.rename({'ID':'SNP'})
    return sumStats
```

```python
fileLocation = '/scratch/eleiterw/AppliedProject/results/'
fileSuffix = '_betterAge_oldEigen_complete.assoc.logistic'
phenotypes = ['remission', 'response', 'wk8Remission', 'wk8Response', 'seasonal']
for pheno in phenotypes:
    filepath = fileLocation + pheno + fileSuffix
    sumStats = formatSumStats(filepath)
    sumStats.to_csv(f'/scratch/eleiterw/AppliedProject/scripts/magma/sumStats/{pheno}.magma.sumstats.txt', index=False, sep = ' ')
```
