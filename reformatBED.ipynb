```python
import pandas as pd
```

```python
bimKeys = pd.read_csv('/home/eleiterw/Thesis/OGFiles/SSRIBIMKeys', sep = '\t')
print(bimKeys.head())
```

```python
bimKeys = bimKeys[['name', '#chrom', 'chromEnd']]
print(bimKeys.head())
```

```python
chromosomes = [str(i) for i in range(1, 23)]
chromosomes.append('X')
print(chromosomes)
```

```python
bimKeys = bimKeys[:][bimKeys['#chrom'].str.replace("chr","").isin(chromosomes)]
print(bimKeys.shape)
```

```python
print(bimKeys[:][bimKeys['name'].duplicated(keep = False)])
```

```python
toRemove = [4541, 303757, 77407, 565658, 402210]
bimKeys = bimKeys.drop(toRemove, axis=0).reset_index(drop=True)
```
